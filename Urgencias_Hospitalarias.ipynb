```python
import pandas as pd
```

#Cargamos los Diccionarios de atención

```python
df_anexo1 = pd.read_excel("Diccionario.xlsx")
df_anexo1
```

```python
causas_resp = [3,4,5,6,7]
comunas = ["Osorno", "Temuco", "Valdivia", "Concepción", "Chiguayante"]
```

#Cargamos los set de datos del 2020 hasta el 2025

```python
df2025 = pd.read_csv("AtencionesUrgencia2025.csv", encoding='latin1', sep=';')
df2024 = pd.read_csv("AtencionesUrgencia2024.csv", encoding='latin1', sep=';')
df2023 = pd.read_csv("AtencionesUrgencia2023.csv", encoding='latin1', sep=';')
df2022 = pd.read_csv("AtencionesUrgencia2022_Mapeado.csv", encoding="utf-8")
df2021 = pd.read_csv("AtencionesUrgencia2021_2_Mapeado.csv", encoding="utf-8")
df2020 = pd.read_csv("AtencionesUrgencia2020_Mapeado.csv", encoding="utf-8")
```

```python
print(df2020["ComunaGlosa"].str.startswith("Concepción").sum())
```

#Filtramos por causas respiratorias

```python
df2020_filtrado = df2020[df2020["IdCausa"].isin(causas_resp) & df2020["ComunaGlosa"].isin(comunas)]
df2021_1_filtrado = df2021[df2021["IdCausa"].isin(causas_resp) & df2021["ComunaGlosa"].isin(comunas)]
df2022_filtrado = df2022[df2022["IdCausa"].isin(causas_resp) & df2022["ComunaGlosa"].isin(comunas)]
df2023_filtrado = df2023[df2023["IdCausa"].isin(causas_resp) & df2023["NombreComuna"].isin(comunas)]
df2024_filtrado = df2024[df2024["IdCausa"].isin(causas_resp) & df2024["NombreComuna"].isin(comunas)]
df2025_filtrado = df2025[df2025["IdCausa"].isin(causas_resp) & df2025["NombreComuna"].isin(comunas)]
```

```python
print(df2020_filtrado["Total"].sum())
print(df2021_1_filtrado["Total"].sum())
print(df2022_filtrado["Total"].sum())
print(df2023_filtrado["Total"].sum())
print(df2024_filtrado["Total"].sum())
print(df2025_filtrado["Total"].sum())
```
